```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
best_score=0.0

for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.6,random_state=i)
    model=LinearRegression()
    model.fit(x_train,y_train)
    s= (model.score(x_test,y_test))
    if(s>best_score):
        best_score=s
        best_state=i
        
print("model score: ",best_score," at random_state: ",best_state)
        

```

```python
best_size=0.0
best_score2=0.0
for i in range(1,10):
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=(i/10),random_state=44)
    model=LinearRegression()
    model.fit(x_train,y_train)
    p=model.score(x_test,y_test)
    if(p>best_score2):
        best_size=i
        best_score2=p
        
print("best train_size is: ",best_size," best score: ",best_score2)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=44)
model=LinearRegression()
model.fit(x_train,y_train)
print("the model r2 score is: ",model.score(x_test,y_test))
```

```python
pred=model.predict(x_test)
```
